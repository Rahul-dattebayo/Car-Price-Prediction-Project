```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
#opening the homepage of flipkart

driver.get('https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId=400001')
```

```python
links = []


url=driver.find_elements_by_xpath('//div[@class="col-4"]/a')#scraping urls

for i in url:

    links.append(i.get_attribute("href"))#appending the urls in links list


#The empty lists for appending it later.

product_Names = []
Years = []
fuels = []
kms_drivens = []
ownerss = []
transmissions = []
prices = []

for i in links:
    driver.get(i)
    time.sleep(1)
    
    try:
        name = driver.find_element_by_xpath("//h2[@class = '_2geSF']")
        product_Names.append(name.text.split(maxsplit=1)[1:])
        
    except:
        product_Names.append('-')
        
    try:
        age = driver.find_element_by_xpath("//ul[@class = '_2KO13']/li[9]/strong[1]")
        Years.append(age.text.split(maxsplit = 1)[1:])
        
    except:
        Years.append('-')
        
    try:
        gas = driver.find_element_by_xpath("//ul[@class = '_2KO13']/li[4]/strong[1]")
        fuels.append(gas.text)
    
    except:
        fuels.append('-')
        
    try:
        dist = driver.find_element_by_xpath("//ul[@class = '_2KO13']/li[3]/strong[1]")
        kms_drivens.append(dist.text.split(' ')[0])
        
    except:
        kms_drivens.append('-')
        
    try:
        own = driver.find_element_by_xpath("//ul[@class = '_2KO13']/li[2]/strong[1]").text.split(' ')[0]
        ownerss.append(own)
        
    except:
        ownerss.append('-')
        
    try:
        gear = driver.find_element_by_xpath("//ul[@class = '_2KO13']/li[6]/strong[1]").text
        transmissions.append(gear)
        
    except:
        transmissions.append('-')
        
    try:
        cost = driver.find_element_by_xpath("//div[@class = 'text-right g-hPT']/strong[1]")
        prices.append(cost.text)
        
    except:
        prices.append('-')
```

```python
cars1 = pd.DataFrame({})
cars1['name'] = product_Names
cars1['year'] = Years
cars1['selling_price'] = prices
cars1['Km_driven'] = kms_drivens
cars1['fuel'] = fuels
cars1['Transmission'] = transmissions
cars1['owner'] = ownerss
```

```python
cars1.sample(5)
```

```python
cars1.to_csv("Cars_Mumbai.csv",index=True)
```
